```python
#normalize dataset to minmax scaler object
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
transformed_df = pd.DataFrame(scaler.fit_transform(data), columns=data.columns)
transformed_df
```

```python
#separating features from label
features_df = transformed_df.drop(columns=["Appliances"])
target_df = transformed_df["Appliances"]
```

```python
features_df.head()
```

```python
target_df.head()
```

```python
#splitting dataset into training and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features_df, target_df, test_size=0.3, random_state = 42)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

```python
#plot training set
plt.scatter(X_train["T_out"], y_train)
```

```python
#creating the linear model
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()

#training the linear regressor model
lin_reg.fit(X_train, y_train)

#predicting test set
lin_y_pred = lin_reg.predict(X_test)
```

```python
#creating a lasso_reg to handle underfitting and overfitting
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)

#traing the lasso regression model
lasso_reg.fit(X_train, y_train)

#predicting test set
lasso_y_pred = lasso_reg.predict(X_test)
```

```python
#creating a ridge_reg to handle underfitting and overfitting
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.5)

#training the ridge reg model
ridge_reg.fit(X_train, y_train)

#predicting the test set
ridge_y_pred = ridge_reg.predict(X_test)
```

```python
#calculating the MSE of each model
from sklearn.metrics import mean_absolute_error

lin_mse = mean_absolute_error(y_test, lin_y_pred)

las_mse = mean_absolute_error(y_test, lasso_y_pred)

rig_mse = mean_absolute_error(y_test, ridge_y_pred)

print("linear mse is: {:.3f}".format(lin_mse), "\nlasso mse is: {:.3f}".format(las_mse), "\nRidge mse is: {:.3f}".format(rig_mse))
```

```python
#calculating the root mean squared
from sklearn.metrics import mean_squared_error

lin_rmse = np.sqrt(mean_squared_error(y_test, lin_y_pred))

las_rmse = np.sqrt(mean_squared_error(y_test, lasso_y_pred))

rig_rmse = np.sqrt(mean_squared_error(y_test, ridge_y_pred))

print("linear rmse is: {:.3f}".format(lin_rmse), "\nlasso rmse is: {:.3f}".format(las_rmse), "\nRidge rmse is: {:.3f}".format(rig_rmse))
```

```python
#calculating the residual sum of squares


lin_rss = np.sum(np.square(y_test - lin_y_pred))

las_rss = np.sum(np.square(y_test - lasso_y_pred))

rig_rss = np.sum(np.square(y_test - ridge_y_pred))

print("linear lin_rss is: {:.3f}".format(lin_rss), "\nlasso las_rss is: {:.3f}".format(las_rss), "\nRidge rig_rss is: {:.3f}".format(rig_rss))
```

```python
#calculating the r squared of each model
from sklearn.metrics import r2_score

lin_r2 = r2_score(y_test, lin_y_pred)

las_r2 = r2_score(y_test, lasso_y_pred)

rig_r2 = r2_score(y_test, ridge_y_pred)

print("linear r2 score is: {:.3f}".format(lin_r2), "\nlasso r2 score is: {:.3f}".format(las_r2), "\nRidge r2 score is: {:.3f}".format(rig_r2))
```

```python
#calculating the weights and observing effects of reglarization

def get_weights_df(model, features, col_name):
    #getting the weights of each model
    weights = pd.Series(model.coef_, features.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ["Features", col_name]
    weights_df[col_name].round(4)
    return weights_df

linear_model_weights = get_weights_df(lin_reg, X_train, "Linear_Model_weights")
ridge_model_weights = get_weights_df(ridge_reg, X_train, "Ridge_Model_Weights")
lasso_model_weights = get_weights_df(lasso_reg, X_train, "Lasso_Model_Weights")


final_weights = pd.merge(linear_model_weights, ridge_model_weights, on="Features")
final_weights = pd.merge(final_weights, lasso_model_weights, on="Features")
```
